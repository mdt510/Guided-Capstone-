```python
# First, import the relevant modules
import requests
import json
```

```python
# Now, call the Quandl API and pull out a small sample of the data (2017) to get a glimpse into the JSON structure that will be returned
url = 'https://www.quandl.com/api/v3/datasets/FSE/AFX_X?start_date=2017-01-01&end_date=2017-12-31&api_key='
r = requests.get(url)
json_data = r.json()

# Inspect the JSON structure of the object you created, and take note of how nested it is, as well as the overall structure
# Print each key-value pair in json_data
for k in json_data.keys():
   print(k + ': ', json_data[k])
```

```python
#make 'json_data' a JSON file-type
data = json.dumps(json_data)

#convert JSON variable to Python dictionary
python_dict = json.loads(data)
print (python_dict)
```

**Calculate what the highest and lowest opening prices were for the stock in this period**

```python
column_values = python_dict['dataset']['data']
def Extract(two):
  return [item[1] for item in two]

open_price = Extract(column_values)


#create a new list for open_price not including the value 'None'
new_open_price = [price for price in open_price if price!=None]

print 'highest opening price:', max(new_open_price)
print 'lowest opening price:', min(new_open_price)
```

**What was the largest change in any one day (based on High and Low price)?**

```python
#create list of 'Highs'
def Extract(three):
  return [item[2] for item in three]

high_price = Extract(column_values)

#create list of 'Lows'
def Extract(four):
  return [item[3] for item in four]

low_price = Extract(column_values)

#create a list of 'amout_change' for each day
price_change=[]

for i in range(len(high_price)):
  price_change.append(high_price[i] - low_price[i])

print 'Max daily price change:', max(price_change)
```

**What was the largest change between any two days (based on Closing Price)?**

```python
#create list of closing prices
def Extract(five):
  return [item[4] for item in five]

closing_price = Extract(column_values)

#create list of daily closing change
closing_change = [closing_price[i + 1] - closing_price[i] for i in range(len(closing_price)-1)]

abs_closing_change =  [abs(ele) for ele in closing_change]

print 'Largest change between days:', max(abs_closing_change)
```

**What was the average daily trading volume during this year?**

```python
#create list of daily trading volume
def Extract(seven):
  return [item[6] for item in seven]

trading_vol = Extract(column_values)

#calculate and print the average daily trading volume
print 'average daily trading volume:', sum(trading_vol)/len(trading_vol)
```
